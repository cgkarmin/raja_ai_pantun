```python
import pandas as pd

# Baca fail dengan mengabaikan baris pertama (header tambahan)
with open("data/submit_pantun.csv", "r", encoding="utf-8") as f:
    lines = f.readlines()

# Buang header tambahan
lines = [line for line in lines if not line.startswith("Nama,Email,No WhatsApp,Pantun")]

# Gabungkan baris pantun dalam satu lajur tanpa tanda koma berlebihan
cleaned_lines = []
for line in lines:
    parts = line.strip().split(",")  # Pecahkan kepada lajur
    nama, email, no_wa = parts[:3]  # Ambil 3 lajur pertama
    pantun = ",".join(parts[3:]).replace(",,,,", "").strip('"')  # Gabungkan pantun & buang koma berlebihan
    cleaned_lines.append([nama, email, no_wa, pantun])

# Simpan semula fail CSV dengan format betul
df_fixed = pd.DataFrame(cleaned_lines, columns=["Nama", "Email", "No WhatsApp", "Pantun"])
df_fixed.to_csv("data/submit_pantun.csv", index=False, encoding="utf-8")

print("‚úÖ Data dalam 'submit_pantun.csv' telah dibersihkan dan diperbaiki!")
```

```python
import pandas as pd

# Baca fail CSV (pastikan fail ini berada dalam folder 'data')
file_path = "data/raja_pantun_no_bom.csv"
df = pd.read_csv(file_path, encoding="utf-8")

# **Tapis pantun dengan markah 23-25 sahaja**
df_filtered = df[(df["markah_penuh"] >= 23) & (df["markah_penuh"] <= 25)]

# **Kira jumlah pantun dan markah keseluruhan setiap pemantun**
leaderboard = df_filtered.groupby("author").agg(
    jumlah_pantun=("text", "count"),
    markah_keseluruhan=("markah_penuh", "sum"),
    purata_markah=("markah_penuh", "mean")
).reset_index()

# **Susun leaderboard berdasarkan markah keseluruhan dan jumlah pantun**
leaderboard = leaderboard.sort_values(by=["markah_keseluruhan", "jumlah_pantun"], ascending=[False, False])

# **Tetapkan gelaran kepada 3 pemantun terbaik**
titles = ["Tanjak Emas", "Tanjak Perak", "Tanjak Suasa"]
leaderboard["Gelaran"] = [titles[i] if i < len(titles) else "" for i in range(len(leaderboard))]

# **Simpan leaderboard ke dalam CSV untuk kegunaan laman web**
leaderboard.to_csv("data/leaderboard_pantun.csv", index=False, encoding="utf-8")

# **Paparkan leaderboard dalam Jupyter Notebook**
import ace_tools as tools
tools.display_dataframe_to_user(name="Leaderboard Pemantun", dataframe=leaderboard)
```
